Exploring different uses of linear regression and polyfit using ridge

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets, linear_model, preprocessing
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.cross_validation import train_test_split
%matplotlib inline
```

```python
# A function that transforms one variable into another variable
def f(x):
    return np.sin(x) + 0.5*x
```

```python
# Generate the series of data
X = pd.DataFrame(np.linspace(-5 * np.pi, 5 * np.pi, 5000))
y = f(X)
```

```python
plt.figure(figsize=(10,5))
plt.plot(X, y)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
# Train Linear Regression Model
regr=linear_model.LinearRegression()
regr.fit(X,y)

# Make Prediction
y_pred=regr.predict(X)
```

```python
plt.figure(figsize=(10,5))
plt.plot(X, y, 'b', label='f(x)')
plt.plot(X, y_pred, 'r-', label='regression')
plt.legend(loc=0)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
poly = make_pipeline(PolynomialFeatures(13), Ridge())
poly.fit(X, y)
y_pred2=poly.predict(X)
#    y_plot = model.predict(X_plot)
plt.figure(figsize=(10,5))
plt.plot(X, y, 'b', label='f(x)')
plt.plot(X, y_pred, 'r-', label='regression')
plt.plot(X, y_pred2, 'g', label='poly')
plt.legend(loc=0)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
```
