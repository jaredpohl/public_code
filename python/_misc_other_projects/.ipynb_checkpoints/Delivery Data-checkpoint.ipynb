```python
df_raw=pd.read_csv('delivery_data.csv')

# add asof_dt
asof_dt=[]
for i in range(0,len(df_raw)):
    asof_dt.append(datetime.strptime(df_raw['asof_dt'].ix[i],'%m/%d/%Y').strftime('%Y-%m-%d'))
df_raw['asof_dt']=pd.to_datetime(asof_dt)

# add weekend flag
wknd=[]
for i in range(0,len(df_raw)):
    d=df_raw['asof_dt'].ix[i].strftime("%w")
    t=['5','6','0']
    if d in t:
        wknd.append('Y')
    else:
        wknd.append('N')
df_raw['wknd']=wknd
df_raw.head(3)
```

```python
#Total dollars
def drv_tot_usd():
    df=pd.DataFrame(df_raw.groupby('shopper_id')['total_bill_usd'].sum())#.columns=['usd']
    df.columns=['tot_usd']
    return df

# days worked
def drv_days_worked():
    df=pd.DataFrame(df_raw.groupby('shopper_id')['asof_dt'].nunique())
    df.columns=['days_worked']
    return df

# number of deliveries
def drv_n_deliveries():
    df=pd.DataFrame(df_raw.groupby('shopper_id')['asof_dt'].count())
    df.columns=['n_deliveries']
    return df

#create summary data
def drv_tot():
    df=pd.concat([drv_tot_usd(),drv_days_worked(),drv_n_deliveries()], axis=1, join='inner')
    df=df.reset_index()
    df['deliveries_pday']=df['n_deliveries']/df['days_worked']
    df['tot_usd_day']=df['tot_usd']/df['days_worked']
    df=df.sort_values('tot_usd', ascending=False).reset_index(drop=True)
    return df
```

```python
# Plotting Histogram with CDF
def plot_hist_cdf(data,n):
    plt.close()
    f, (ax1, ax2) = plt.subplots(1, 2, sharey=False,)
    ax1.hist(data, bins=n)
    ax1.set_title('Histogram')
    ax2.hist(data, bins=20, cumulative=True,histtype='step', color='Black')
    ax2.set_title('Cumulative')
    f.set_figwidth(15)
    fig = plt.gcf()

# Linear Regression with Plot
def linreg(X,Y, xlab, ylab):
    # Running the linear regression
    X = sm.add_constant(X)
    model = regression.linear_model.OLS(Y, X).fit()
    a = model.params[0]
    b = model.params[1]
    X = X[:, 1]
    # Return summary of the regression and plot results
    X2 = np.linspace(X.min(), X.max(), 100)
    Y_hat = X2 * b + a
    plt.scatter(X, Y, alpha=0.3) # Plot the raw data
    plt.plot(X2, Y_hat, 'r', alpha=0.9);  # Add the regression line, colored in red
    plt.xlabel(xlab)
    plt.ylabel(ylab)
#    print('Model Parameters:',model.params)
#    print('Model pValues:',model.pvalues)
    return 
```

### Observations:
We begin by aggregating the data by driver, and finding the total amount of money they made, total days each driver worked, number of deliveries they gave and deliveries per day. Here are some observations:

**Total USD:** Sorted by total dollars made by driver during this time period, we can see the drivers sorted by dollars made

**Total Days:** We can observe that the total number of days worked follows a similar path as the dollars made in the top half of the drivers. The bottom half of drivers tend to drive sporadically. We begin to see a dilineation of driver discipline - some drivers drive everyday, while others drive sporadically.

**Deliveries:** Total number of deliveries given is similar to total USD. This is as expected.

**Deliveries Per Day:**  We see a relatively consistent deliveries per day in the top half. In the bottom half, we see sporadic deliveries per day for each drive. This suggests that the disciplined drivers drive consistently throughout the day, while the sporadic drivers are inconsistent throughout the day.

We can make an observation that there are two types of drivers:
* Drivers who are consistent on a daily basis tend to be drivers who are disciplined throughout the day.
* Drivers who drive sporadically tend to be less disciplined throughout the day.

### Modeling Total Dollars Made 

We can regress the total dollars made by each driver throughout the period by both the total days worked and the total deliveries per day. We can observe that both values are significant (low P value).

```python
# Use column_stack to combine independent variables, then add a column of ones so we can fit an intercept

X1=df_tot.days_worked.values
X2=df_tot.deliveries_pday.values

Y=df_tot.tot_usd
X = sm.add_constant( np.column_stack( (X1, X2) ) )

# Run the model
results = regression.linear_model.OLS(Y, X).fit()

print ('Params:', results.params)
print ('p_values:', results.pvalues)
print(results.summary2())
```

From this regression model it appears that for each additional day the driver drives, he/she will make 50 dollars more per day over the period. For each additional deliveries the driver gives per day, the driver will make an additional $190 dollars over the period
